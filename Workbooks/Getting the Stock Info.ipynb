```python
other=["MSFT", "AAPL", "AMD", "DIS", "QQQ", "V", "BA", "TSLA", "AMZN", "VOO", "FB", "RTX", "DKNG", "KO", "NVDA", "T", "SQ", "BABA", "JPM", "DAL", "CCL", "MGM", "NIO", "VTI", "NET", "SE", "XOM", "MA", "BRK-B", "VGT", "WMT"]
```

```python
#Json tracker
def track_info(x):
    url = "https://apidojo-yahoo-finance-v1.p.rapidapi.com/stock/get-detail"

    querystring = {"symbol":x}

    headers = {
    'x-rapidapi-key': "eb056a72f9msh2564512806e6aebp12a874jsnd03d93885916",
    'x-rapidapi-host': "apidojo-yahoo-finance-v1.p.rapidapi.com"
    }

    response = requests.request("GET", url, headers=headers, params=querystring)
    results=response.json()

    return results
```

```python
#ticker
Ticker=other

#Name
Name=[]
for x in other:
    try:
        results=track_info(x)
        Name.append(results["price"]["shortName"])
    except:
        Name.append(" ") 

#Exchange
Exchange=[]
for x in other:
    try:
        results=track_info(x)
        Exchange.append(results["price"]["exchange"])
    except:
        Exchange.append(" ") 



#Country
Country=[]
for x in other:
    try:
        results=track_info(x)
        Country.append(results["summaryProfile"]["country"])
    except:
        Country.append(" ") 

#Sector
Sector=[]
for x in other:
    try:
        results=track_info(x)
        Sector.append(results["summaryProfile"]["sector"])
    except:
        Sector.append(" ") 

#Industry
Industry=[]
for x in other:
    try:
        results=track_info(x)
        Industry.append(results["summaryProfile"]["industry"])
    except:
        Industry.append(" ")

#Currency
Currency=[]

for x in other:
    try:
        results=track_info(x)
        Currency.append(results["earnings"]["financialCurrency"])
    except:
        Currency.append(" ") 

#Market
Market=[]

for x in other:
    try:
        Market.append(results["quoteType"]["market"])
    except:
        Market.append(" ")

        
#Nber_of_analyst_opinions
Nber_of_analyst_opinions=[]

for x in other:
    try:
        results=track_info(x)
        Nber_of_analyst_opinions.append(results["financialData"]["numberOfAnalystOpinions"]["raw"])
    except:
        Nber_of_analyst_opinions.append(" ")

#MarketCap        
MarketCap=[]

for x in other:
    try:
        results=track_info(x)
        MarketCap.append(results["price"]["marketCap"]["raw"])
    except:
        MarketCap.append(" ")

#Volatility
Volatility=[]

for x in other:
    try:
        results=track_info(x)
        Volatility.append(results["defaultKeyStatistics"]["beta"]["raw"])
    except:
        Volatility.append(" ")

#Recommendation_mean
Recommendation_mean=[]

for x in other:
    try:
        results=track_info(x)
        Recommendation_mean.append(results["financialData"]["recommendationMean"]["raw"])
    except:
        Recommendation_mean.append(" ")
```
